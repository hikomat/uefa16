```python
import pandas as pd
import requests
```

```python
def create_livescore_endpoint(x):
    year, month, day = x['dtf']
    matchid = x['matchid']
    url = 'http://www.uefa.com/live/match-centre/livescores/'
    params = 'year={}/month={}/day={}/match={}/livescore.json'.format(year, month, day, matchid)
    return url + params
```

```python
df = pd.read_csv('matches.csv', header=None)
df.columns = ['h', 'v','date','link']
df['matchid'] = df.link.map(lambda x: x.split('/')[-2][-7:])
df['dt'] = pd.to_datetime(df.date)
df['dtf'] = df['dt'].map(lambda x: (x.year, x.month, x.day))
df['livescore_endpoint'] = df.apply(create_livescore_endpoint, axis=1)
df['daas_req'] = df.matchid.map(lambda x: 'http://daaseuro2016.uefa.com/api/v2/football/en/matches/{}'.format(x))
```
